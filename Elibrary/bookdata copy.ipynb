```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
소설 ='http://www.kyobobook.co.kr/bestSellerNew/bestseller.laf?mallGb=KOR&linkClass=B&range=1&kind=3&orderClick=DAf'
```

```python
# 교보문고의 베스트셀러 웹페이지를 가져옵니다.
html = urlopen('http://www.kyobobook.co.kr/bestSellerNew/bestseller.laf?mallGb=KOR&linkClass=d&range=1&kind=3&orderClick=DAf')
bsObject = BeautifulSoup(html, "html.parser")

title =[]
date=[]
bookcode=[]
author=[]
name=[]
genre=[]
소개=[]
서평=[]
image=[]
book_page_urls = []
for cover in bsObject.find_all('div', {'class':'detail'}):
    link = cover.select('a')[0].get('href')
    book_page_urls.append(link)
```

```python
bsObject.find_all('div', {'class':'detail'})
```

```python
# 메타 정보로부터 필요한 정보를 추출합니다.메타 정보에 없는 저자 정보만 따로 가져왔습니다.  
for index, book_page_url in enumerate(book_page_urls):
    html = urlopen(book_page_url)
    bsObject = BeautifulSoup(html, "html.parser")


    try:
        title.append(bsObject.find('meta', {'property':'og:title'}).get('content').replace(' - 교보문고',""))#제목
        date.append(bsObject.find('span','date').get_text().strip().replace(' 출간',""))#출판일
        bookcode.append(index+1) #책코드
        author.append(bsObject.select('span.name a')[0].text.strip()) #지은이
        try:#출판사
            name.append(bsObject.find_all('span','name')[1].get_text().strip())
        except IndexError as e:
            name.append(bsObject.find_all('span','name')[2].get_text().strip())
        genre.append('여행') #장르
        소개.append(bsObject.find_all('div','content')[1].get_text().strip()) #작품소개
        try:
            서평.append(bsObject.find_all('div','content')[3].get_text().strip()) #서평 
        except IndexError as e:
            서평.append("등록된 서평 없음")
        image.append(bsObject.find('meta', {'property':'og:image'}).get('content')) #이미지주소
        print(index+1)
    except:
        pass
```

```python
df = pd.DataFrame([ x for x in zip(title,date,bookcode,author,name,genre,소개,서평,image)])
df
```

```python
df.to_excel('시더미데이터.xlsx')
```
